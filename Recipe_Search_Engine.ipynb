```
all_querys = []
for i in range(len(query)):
  tokenizer = nltk.RegexpTokenizer(r'\w+')
  prova = tokenizer.tokenize(query[i].lower())
  lemmatizer = WordNetLemmatizer()
  query_new = []
  for i in prova:
    query_new.append(lemmatizer.lemmatize(i))
  
  all_querys.append(query_new)

print(all_querys)
```

```
ingredients = df['list_of_ingredients']
```

```
# Tokenizing the list of ingredients
stop = set(stopwords.words('english'))
tok_ingr = []
for sentence in ingredients[:-1]:
  line = word_tokenize(sentence)
  tok_ingr.append(line)

# remove punctuation and lower letters
tokens = []
for row in tok_ingr:
  # print(row)
  new_row = []
  for i in range(len(row)):
    word = row[i]
    if word != '[' and word != ']' and word != ',' and word != ')' and word != '(':
      if word.isalpha:
        word = ''.join([i for i in word if not i.isdigit()])
        if word != 'kg' and word != 'g'and word != '' and word not in stop and word!= 'tsp' and word!= 'tbsp' and word!= 'salt' and word!= 'pepper' and word != '½' and word != 'ml' and word != '’':
          new_row.append(lemmatizer.lemmatize(word))
  tokens.append(new_row)
  
print(tokens[1])
```

```
from collections import Counter
#UNIGRAM - count term frequency and have a cut off point (we can decide)
unigram = Counter()
for l in tokens:
  for ing in l:
    unigram[ing] += 1;

unigram_total = sum(unigram.values()) # to get the denominator for unigram probabilities
print (len(unigram), "different unigrams observed")
print("unigram total", unigram_total)
print(unigram)
```

### Precision = No. of relevant documents retrieved / No. of total documents retrieved
No. of relevant documents retrieved  = top 5 that we are returning (and if they have a socre of non zero)
No. of total documents retrieved = all document with non-zero score

```
all_recipe_scores = []
all_top_5 = []
all_precision = []
all_recall = []
total_recipe_non_zero = []
total_recipe_greater_one = []
for q in all_querys:
  recipe_scores = bm25.get_scores(q)
  all_recipe_scores.append(recipe_scores)
  top_5 = bm25.get_top_n(q, tokens, n=10)
  # print(top_5)
  all_top_5.append(top_5)
  query_non_zero = []
  query_greater_1 = []
  for i in range(len(recipe_scores)):
    if recipe_scores[i]!=0:
      query_non_zero.append(recipe_scores[i])
      if recipe_scores[i]>1:
        query_greater_1.append(recipe_scores[i])
  total_recipe_non_zero.append(query_non_zero)
  total_recipe_greater_one.append(query_greater_1)

for q in range(len(all_querys)):
  if total_recipe_greater_one[q] != []:
    precision = 10/len(total_recipe_non_zero[q])
  else:
      precision = 0
  all_precision.append(precision)
  
  if total_recipe_greater_one[q] != []:
    recall = 10/len(total_recipe_greater_one[q])
  else:
      recall = 1
  all_recall.append(recall)

# total_recipe_non_zero = total_recipe_non_zero.sort()
# top_5_score = total_recipe_non_zero[:5]

# relevant_recipe_non_zero = 0
# for i in top_5_score:
#   if i!= 0:
#     relevant_recipe_non_zero += 1
```
